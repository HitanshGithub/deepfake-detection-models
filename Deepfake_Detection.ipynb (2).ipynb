```python
# 1. Imports
import os
import torch
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
from torchvision.models import mobilenet_v2
from sklearn.metrics import confusion_matrix, classification_report
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

# 2. Paths
train_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Train'
val_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Validation'
test_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Test'
model_save_path = "deepfake_mobilenetv2.pth"

# 3. Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("ðŸ”§ Using device:", device)

# 4. Transforms (with Augmentation for Training)
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# 5. Load Datasets
train_data = ImageFolder(train_dir, transform=train_transform)
val_data = ImageFolder(val_dir, transform=test_transform)
test_data = ImageFolder(test_dir, transform=test_transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32)
test_loader = DataLoader(test_data, batch_size=32)

# 6. Compute Class Weights
class_counts = Counter(train_data.targets)
weights = [1.0 / class_counts[i] for i in range(len(class_counts))]
weights = torch.FloatTensor(weights).to(device)

# 7. Load Pretrained Model
model = mobilenet_v2(pretrained=True)
for param in model.features.parameters():
    param.requires_grad = True  # fine-tune entire model
model.classifier[1] = nn.Linear(model.last_channel, 2)
model = model.to(device)

# 8. Loss & Optimizer
criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = optim.Adam(model.parameters(), lr=0.0003, weight_decay=1e-5)

# 9. Training Loop with Validation
epochs = 10
best_val_acc = 0
patience = 3
patience_counter = 0

train_losses, val_losses = [], []
train_accuracies, val_accuracies = [], []

for epoch in range(epochs):
    model.train()
    running_loss, correct, total = 0.0, 0, 0

    print(f"\nðŸš€ Epoch {epoch + 1}/{epochs}")
    for batch_idx, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        if (batch_idx + 1) % 10 == 0 or (batch_idx + 1) == len(train_loader):
            print(f"  ðŸŸ© Batch {batch_idx+1}/{len(train_loader)} - Loss: {loss.item():.4f}")

    train_acc = 100 * correct / total
    train_losses.append(running_loss)
    train_accuracies.append(train_acc)

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item()
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    val_acc = 100 * val_correct / val_total
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)

    print(f"ðŸ“Š Epoch Summary: Train Acc: {train_acc:.2f}%, Train Loss: {running_loss:.2f} | Val Acc: {val_acc:.2f}%, Val Loss: {val_loss:.2f}")

    # Early stopping
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        patience_counter = 0
        torch.save(model.state_dict(), model_save_path)
        print("âœ… Model improved and saved.")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("â¹ï¸ Early stopping triggered.")
            break

# 10. Load Best Model
model.load_state_dict(torch.load(model_save_path))

# 11. Plot Accuracy & Loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_accuracies, label='Train Acc')
plt.plot(val_accuracies, label='Val Acc')
plt.title('Accuracy per Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.title('Loss per Epoch')
plt.legend()
plt.tight_layout()
plt.show()

# 12. Evaluation on Test Set
model.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

# 13. Report
print("\nðŸ“Š Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
print("\nðŸ“‹ Classification Report:\n", classification_report(y_true, y_pred, target_names=train_data.classes))
```

```python
# 1. Imports
import os
import torch
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder  # âœ… Required for loading dataset
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
import timm
from sklearn.metrics import confusion_matrix, classification_report
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# 2. Paths
train_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Train'
val_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Validation'
test_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Test'
model_save_path = "deepfake_coat_model.pth"

# 3. Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("ðŸ”§ Using device:", device)

# 4. Transforms
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# 5. Datasets & Loaders
train_data = ImageFolder(train_dir, transform=train_transform)
val_data = ImageFolder(val_dir, transform=test_transform)
test_data = ImageFolder(test_dir, transform=test_transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32)
test_loader = DataLoader(test_data, batch_size=32)

# 6. Class Weights
class_counts = Counter(train_data.targets)
weights = [1.0 / class_counts[i] for i in range(len(class_counts))]
weights = torch.FloatTensor(weights).to(device)

# 7. Load CoaT Model (using timm)
model = timm.create_model('coat_lite_mini', pretrained=True)  # Using CoaT Lite Mini model
model.head = nn.Linear(model.head.in_features, 2)  # âœ… Modify classifier head for binary classification
model = model.to(device)

# 8. Loss & Optimizer
criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = optim.Adam(model.parameters(), lr=0.0003, weight_decay=1e-5)

# 9. Training Loop
epochs = 10
best_val_acc = 0
patience = 3
patience_counter = 0

train_losses, val_losses = [], []
train_accuracies, val_accuracies = [], []

for epoch in range(epochs):
    model.train()
    running_loss, correct, total = 0.0, 0, 0
    print(f"\nðŸš€ Epoch {epoch + 1}/{epochs}")

    for images, labels in tqdm(train_loader, desc=f"Training Epoch {epoch+1}"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    train_acc = 100 * correct / total
    train_losses.append(running_loss)
    train_accuracies.append(train_acc)

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc=f"Validation Epoch {epoch+1}"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item()
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    val_acc = 100 * val_correct / val_total
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)

    print(f"ðŸ“Š Train Acc: {train_acc:.2f}%, Train Loss: {running_loss:.2f} | Val Acc: {val_acc:.2f}%, Val Loss: {val_loss:.2f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        patience_counter = 0
        torch.save(model.state_dict(), model_save_path)
        print("âœ… Model improved and saved.")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("â¹ï¸ Early stopping triggered.")
            break

# 10. Load Best Model
model.load_state_dict(torch.load(model_save_path))

# 11. Plot Accuracy & Loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_accuracies, label='Train Acc')
plt.plot(val_accuracies, label='Val Acc')
plt.title('Accuracy per Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.title('Loss per Epoch')
plt.legend()
plt.tight_layout()
plt.show()

# 12. Test Evaluation
model.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for images, labels in tqdm(test_loader, desc="ðŸ” Evaluating on Test Set"):
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

# 13. Report
print("\nðŸ“Š Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
print("\nðŸ“‹ Classification Report:\n", classification_report(y_true, y_pred, target_names=train_data.classes))
```

```python
import torch
import torchvision.transforms as transforms
from torchvision import models, datasets
import timm
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import os
from tqdm import tqdm

# Set up device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Model configurations - MODIFIED TO HANDLE BINARY CLASSIFICATION
MODELS = {
    'MobileNetV2': {
        'path': 'deepfake_mobilenetv2.pth',
        'loader': lambda: models.mobilenet_v2(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'MobileNetV3': {
        'path': 'deepfake_mobilenetv3.pth',
        'loader': lambda: models.mobilenet_v3_small(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'ResNet18': {
        'path': 'deepfake_resnet18.pth',
        'loader': lambda: models.resnet18(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'EfficientNetB0': {
        'path': 'deepfake_efficientnet_b0.pth',
        'loader': lambda: models.efficientnet_b0(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'CoaT-Lite': {
        'path': 'deepfake_coat_model.pth',
        'loader': lambda: timm.create_model('coat_lite_mini', pretrained=False, num_classes=2),
        'modifier': lambda m: m
    }
}

# Image transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def load_models():
    """Load all models with error handling"""
    loaded_models = {}
    for name, config in MODELS.items():
        try:
            model = config['loader']()
            config['modifier'](model)
            # Load state dict with strict=False to handle architecture differences
            model.load_state_dict(torch.load(config['path'], map_location=device), strict=False)
            model.to(device).eval()
            loaded_models[name] = model
            print(f"âœ… {name} loaded successfully")
        except Exception as e:
            print(f"âŒ Failed to load {name}: {str(e)}")
    return loaded_models

def evaluate_model(model, test_loader):
    """Evaluate a single model on the test set"""
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc=f"Evaluating {model.__class__.__name__}"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return correct / total

def main():
    # Load all models
    models = load_models()
    if not models:
        print("No models loaded successfully")
        return

    # Load test dataset
    test_dir = r'C:\Users\hitan\Downloads\archive (3)\Dataset\Test'
    test_dataset = datasets.ImageFolder(test_dir, transform=transform)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    print(f"\nEvaluating models on test set ({len(test_dataset)} images)...")
    
    # Evaluate each model
    accuracies = {}
    for name, model in models.items():
        accuracy = evaluate_model(model, test_loader)
        accuracies[name] = accuracy
        print(f"{name}: Test Accuracy = {accuracy:.2%}")
    
    # Display results
    plt.figure(figsize=(10, 5))
    plt.bar(accuracies.keys(), accuracies.values())
    plt.title('Model Test Accuracies')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    for i, v in enumerate(accuracies.values()):
        plt.text(i, v + 0.02, f"{v:.2%}", ha='center')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import torch
import torchvision.transforms as transforms
from torchvision import models
from PIL import Image
import timm
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Set up device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Model configurations - MODIFIED TO HANDLE BINARY CLASSIFICATION
MODELS = {
    'MobileNetV2': {
        'path': 'deepfake_mobilenetv2.pth',
        'loader': lambda: models.mobilenet_v2(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'MobileNetV3': {
        'path': 'deepfake_mobilenetv3.pth',
        'loader': lambda: models.mobilenet_v3_small(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'ResNet18': {
        'path': 'deepfake_resnet18.pth',
        'loader': lambda: models.resnet18(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'EfficientNetB0': {
        'path': 'deepfake_efficientnet_b0.pth',
        'loader': lambda: models.efficientnet_b0(pretrained=False, num_classes=2),
        'modifier': lambda m: m  # No modification needed
    },
    'CoaT-Lite': {
        'path': 'deepfake_coat_model.pth',
        'loader': lambda: timm.create_model('coat_lite_mini', pretrained=False, num_classes=2),
        'modifier': lambda m: m
    }
}

# Image transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def load_models():
    """Load all models with error handling"""
    loaded_models = {}
    for name, config in MODELS.items():
        try:
            model = config['loader']()
            config['modifier'](model)
            # Load state dict with strict=False to handle architecture differences
            model.load_state_dict(torch.load(config['path'], map_location=device), strict=False)
            model.to(device).eval()
            loaded_models[name] = model
            print(f"âœ… {name} loaded successfully")
        except Exception as e:
            print(f"âŒ Failed to load {name}: {str(e)}")
    return loaded_models

def predict_image(model, image_tensor):
    """Make prediction for a single model"""
    with torch.no_grad():
        outputs = model(image_tensor)
        probs = torch.softmax(outputs, dim=1)
        conf, pred = torch.max(probs, 1)
    return pred.item(), conf.item()

def main():
    # Load all models
    models = load_models()
    if not models:
        print("No models loaded successfully")
        return

    # Load and process image
    image_path = r"C:\Users\hitan\Downloads\archive (3)\Dataset\Validation\Fake\fake_1000.jpg"
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = transform(img).unsqueeze(0).to(device)
    except Exception as e:
        print(f"âŒ Failed to load image: {str(e)}")
        return

    # Get predictions from all models
    results = []
    print("\nModel Predictions:")
    for name, model in models.items():
        pred, conf = predict_image(model, img_tensor)
        result = ('Fake', conf) if pred == 0 else ('Real', conf)
        results.append((name, result[0], result[1]))
        print(f"{name}: {result[0]} (confidence: {result[1]:.2%})")

    # Display results
    plt.figure(figsize=(10, 6))
    plt.imshow(np.array(img))
    plt.title("Model Predictions:\n" + 
             "\n".join(f"{n}: {p} ({c:.2%})" for n, p, c in results),
             fontsize=10)
    plt.axis('off')

    # Calculate majority vote
    votes = [r[1] for r in results]
    majority = Counter(votes).most_common(1)[0]
    print(f"\nFinal Decision: {majority[0]} (agreed by {majority[1]}/{len(results)} models)")
    
    plt.show()

if __name__ == "__main__":
    main()
```
